This transformation is foundational for conducting temporal drift detection across laundering typologies. Tracking typology behavior over time is essential in anti-money laundering (AML), as criminal tactics evolve and certain fraud patterns may emerge or disappear.

By engineering Year and Month fields:

- The dissertation gains the ability to quantify typology trends

### Block 9: Handle Outliers & Cap Amounts


Generates a boxplot to analyze outliers in the Amount feature and saves the image.



```python
#  Winsorize extreme amounts
Q1 = df['Amount'].quantile(0.01)
Q3 = df['Amount'].quantile(0.99)
df['Amount'] = np.clip(df['Amount'], Q1, Q3)
```

This step implements Winsorization, a robust preprocessing technique where extreme values are capped at high/low percentiles, safeguarding models from volatility while preserving key patterns.

Controlling for outliers ensures that extreme values do not disproportionately influence GAN training or classifier decision boundaries. This preprocessing enhances stability, improves representational fidelity, and supports the realism objective of the dissertation’s synthetic data generation framework.

### Block 13: Encode Categoricals

Applies label encoding to all categorical columns using sklearn’s LabelEncoder. This step transforms object-type string features into numeric representations—essential for feeding data into machine learning models and CGAN training.

```python
#  Apply label encoding (one-hot for low cardinality if needed)
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for col in cat_cols:
    df[col] = encoder.fit_transform(df[col].astype(str))
```

Interpretation of Code:

The script loops over all identified categorical columns (cat_cols) and transforms them into integer labels. This is a quick and memory-efficient alternative to one-hot encoding, especially suitable for high-cardinality features. Converting all object types to numeric form ensures model compatibility while preserving category distinctions.

How this adds value to the dissertation

Enables GAN Compatibility: GANs and most ML models require numeric inputs—this step ensures all features are numerically encoded.

Preserves Feature Identity: Label encoding retains category distinctions while avoiding dimensionality explosion.

Prepares for Conditional Synthesis: Encoded categorical variables are now ready to be used in CGAN’s conditional vector (e.g., typology_code, payment_type).

Supports Downstream Modeling: Standardizes preprocessing pipeline for classifiers and other learning algorithms.

### Block 15: XGBoost – Feature Importance & Classifier Diagnostics


This block trains a lightweight XGBoost classifier on the cleaned dataset to:

Assess baseline model performance before any augmentation

Identify top predictive features for laundering detection

Store results for feature engineering and CGAN input refinement



```python
#  Import model and utilities
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import pandas as pd

#  Remove target, datetime, and leakage-prone 'typology_code'
X = df.drop(columns=['Is_laundering', 'Date', 'typology_code'])
y = df['Is_laundering']

# Stratified split to preserve laundering ratio
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

#  Train XGBoost classifier
model = XGBClassifier(eval_metric="logloss", use_label_encoder=False)
model.fit(X_train, y_train)

#  Feature importances
importances = pd.Series(model.feature_importances_, index=X.columns)
importances_sorted = importances.sort_values(ascending=False)

#  Plot top 10 features
plt.figure(figsize=(10, 5))
importances_sorted[:10].plot(kind='barh', color='darkorange')
plt.title("XGBoost – Top 10 Feature Importances (No Leakage)")
plt.xlabel("Feature Importance Score")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig("../outputs/feature_importance_no_leakage.png")
plt.show()

#  Classification report
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

Insight: Payment_type and Amount are the most influential predictors, reinforcing their importance in both real and synthetic training setups.

Critical Weakness: The model barely detects laundering cases (recall = 9%). This is a classic false negative risk in fraud detection and demands serious remediation via class balancing or data augmentation.

How This Adds Value to the Dissertation

- Establishes Baseline - Creates a reproducible benchmark to compare future CGAN-based augmentation and SMOTE sampling
- Feature Engineering Insight - Validates that features like Payment_type, Amount, and Currency are high-signal variables, making them prime candidates for GAN input and SHAP-based explainability
- Justifies Need for CGAN - Demonstrates that conventional classifiers underperform in recall due to extreme class imbalance—highlighting the need for advanced data synthesis
- Supports Per-Typology Strategies - Encourages stratified modeling and typology-specific attention in downstream phases
- Aids Interpretability Planning - Top features serve as anchors for SHAP analysis and fairness audits in later phases
