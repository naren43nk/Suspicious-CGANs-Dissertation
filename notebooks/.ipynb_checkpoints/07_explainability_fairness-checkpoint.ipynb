### Block 2: Load and Extract Valid Features


```python
# Load real and synthetic datasets
real_path = os.path.join(data_dir, "clean_suspicious.csv")
synth_path = os.path.join(data_dir, "final_synthetic_typology_data.csv")

real_df = pd.read_csv(real_path)
synth_df = pd.read_csv(synth_path)

# Create augmented dataset (real + synthetic)
aug_df = pd.concat([real_df, synth_df], ignore_index=True)

print(f"Real dataset size: {len(real_df)}")
print(f"Synthetic dataset size: {len(synth_df)}")
print(f"Augmented dataset size: {len(aug_df)}")

# Identify valid features (exclude target/meta columns)
meta_cols = ["typology_code", "source", "is_suspicious", "transaction_id", "uuid", "date"]
selected_features = [
    col for col in aug_df.columns
    if col not in meta_cols and np.issubdtype(aug_df[col].dtype, np.number)
]

print("Selected Features for Explainability/Fairness:", selected_features)

# Feature & label splits
X_real, y_real = real_df[selected_features], real_df[target_col]
X_aug, y_aug = aug_df[selected_features], aug_df[target_col]

# Train baseline XGBoost classifiers
xgb_real = xgb.XGBClassifier(
    n_estimators=100,
    use_label_encoder=False,
    eval_metric="mlogloss",
    random_state=42
)
xgb_real.fit(X_real, y_real)

xgb_aug = xgb.XGBClassifier(
    n_estimators=100,
    use_label_encoder=False,
    eval_metric="mlogloss",
    random_state=42
)
xgb_aug.fit(X_aug, y_aug)

print("XGBoost models trained: Real-only and Augmented (Real+Synthetic).")
```

In this block, we load the **real** dataset (`clean_suspicious.csv`) and the **synthetic** dataset generated by CGAN (`final_synthetic_typology_data.csv`).  
These are then combined to form the **augmented dataset**, which merges both real and synthetic records.

---

#### Dataset Sizes
- **Real dataset:** 1,048,575 records  
- **Synthetic dataset:** 10,000 records  
- **Augmented dataset:** 1,058,575 records  

This setup ensures that downstream explainability and fairness analysis can be performed on models trained under different data regimes.

---

#### Feature Selection
From the augmented dataset, valid features were automatically extracted by excluding target/meta columns.  
The selected features are: ['Time', 'Sender_account', 'Receiver_account', 'Amount', 'Is_laundering', 'Year', 'Month']


These features represent a mix of **transaction metadata** and **derived attributes** from earlier preprocessing.

---

#### Model Training
Two baseline **XGBoost classifiers** were trained for Phase 7 explainability:  
1. **Real-only model** → trained on the original real dataset.  
2. **Augmented model** → trained on the combined real + synthetic dataset.  

Both models will later be evaluated with **SHAP-based feature attributions** and subjected to **fairness analysis** across typologies.

---

#### Notes

- Both models are now ready for downstream explainability and fairness comparisons.


### Block 3: SHAP on Real-Only Model

```python
# Initialize SHAP explainer for the real-only XGBoost model
explainer_real = shap.TreeExplainer(xgb_real)

# Compute SHAP values for the real dataset
print("Computing SHAP values for Real-only model...")
shap_values_real = explainer_real.shap_values(X_real)

# Save SHAP values for reuse
shap_real_path = os.path.join(results_dir, "shap_values_real.npy")
np.save(shap_real_path, shap_values_real, allow_pickle=True)

print(f"SHAP values (Real-only model) computed and saved to {shap_real_path}")
```

In this block, we apply **SHAP (SHapley Additive exPlanations)** to the **real-only XGBoost model** trained in Block 2.  
The goal is to understand how individual features contribute to predictions when the classifier is trained exclusively on the real dataset.

---

#### Purpose
- Provide a **baseline explainability profile** for AML typology detection using real data.  
- Quantify the **marginal contribution of each feature** (e.g., transaction amount, sender/receiver accounts, laundering flag) towards the model’s classification decisions.  
- Establish a reference against which augmented (real + synthetic) models can later be compared.

---

#### Method
1. **TreeExplainer** (`shap.TreeExplainer`) is used, as it is optimized for tree-based models like XGBoost.  
2. SHAP values are computed across all samples in the real dataset (`X_real`).  
3. The SHAP attributions are stored in `../results/shap_values_real.npy` for reuse in later blocks (e.g., global feature importance plots, typology-specific SHAP heatmaps).

---

#### Why This Matters
- SHAP provides a **transparent lens** into how AML classifiers weigh transaction features.  
- By starting with the **real-only model**, we establish a benchmark explanation set, ensuring that any differences observed in augmented models can be directly attributed to the influence of synthetic data.

### Block 6: SHAP Δ (Delta) Comparison Bar Plot

```python
# Extract SHAP values for one class (e.g., class 0)
shap_real_class0 = shap_values_real[:, :, 0]
shap_aug_class0 = shap_values_aug[:, :, 0]

# Mean absolute SHAP value per feature
shap_mean_real = np.abs(shap_real_class0).mean(axis=0)
shap_mean_aug = np.abs(shap_aug_class0).mean(axis=0)

#  Delta SHAP per feature
delta_shap = shap_mean_aug - shap_mean_real

delta_df = pd.DataFrame({
    'Feature': X_real.columns.tolist(),
    'Real_Only': shap_mean_real,
    'Real_CGAN': shap_mean_aug,
    'Delta': delta_shap
}).sort_values(by='Delta', key=abs, ascending=False)

# Plot Delta
plt.figure(figsize=(10, 6))
sns.barplot(x='Delta', y='Feature', data=delta_df, palette='coolwarm')
plt.title("Δ SHAP (Real+CGAN - Real Only)")
plt.axvline(0, color='black', linestyle='--')
plt.tight_layout()
plt.savefig(f"{results_dir}/shap_delta_barplot.png")
plt.show()
plt.close()
```

This block quantifies how feature importance changes when the model is trained with CGAN-augmented data. It does this by computing:

\[
\Delta_{SHAP} = \text{mean}(|SHAP_{CGAN}|) - \text{mean}(|SHAP_{Real}|)
\]

Each feature’s mean absolute SHAP value is calculated for both the real-only and the augmented model. The difference is then plotted as a horizontal bar chart.

**Plot Interpretation:**
- Positive values (blue): Features became more influential in the augmented model.
- Negative values (red): Features became less influential after augmentation.
- The feature `Is_laundering` shows a large increase in importance, suggesting a shift in model focus when synthetic data is introduced.

*Output:*
- `shap_delta_barplot.png`

This delta plot is a key diagnostic to assess explainability drift caused by CGAN-based data augmentation.

### Block 8:  UMAP on Typology SHAP Profiles

```python
import umap.umap_ as umap

# UMAP Projection of typology SHAP vectors
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(typology_shap_matrix.values)

# Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(embedding[:, 0], embedding[:, 1], s=100, c=typology_shap_matrix.index, cmap='tab20')
for i, label in enumerate(typology_shap_matrix.index):
    plt.text(embedding[i, 0]+0.1, embedding[i, 1], str(label), fontsize=9)

plt.title("Typology SHAP Profiles via UMAP")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.tight_layout()
plt.savefig(f"{results_dir}/umap_typology_shap.png")
plt.show()
plt.close()
```

To better understand the **similarity in feature attribution across typologies**, this block applies Uniform Manifold Approximation and Projection (UMAP) to the typology-wise SHAP matrix.

- Each typology is represented as a point in 2D space.
- The closer two points are, the more similar their explanation profiles (SHAP vectors).

**Key Observations:**
- Typologies `5`, `6`, `7`, `9`, and `14` form a tight cluster, suggesting they are similarly treated by the CGAN-augmented model.
- Typologies `15`, `17`, `20`, and `23` are more spread, potentially indicating diverse laundering patterns.

### Block 11: Define Fairness Metric Functions


```python
from sklearn.metrics import confusion_matrix

# Demographic Parity: P(ŷ = t | typology = t)
def demographic_parity(y_true, y_pred, typologies):
    results = {}
    for t in np.unique(typologies):
        idx = (typologies == t)
        results[t] = (y_pred[idx] == t).mean()
    return results

# Equal Opportunity: TPR = TP / (TP + FN) for each typology
def equal_opportunity(y_true, y_pred, typologies):
    tpr = {}
    for t in np.unique(typologies):
        idx = (typologies == t)
        y_true_bin = (y_true == t)
        y_pred_bin = (y_pred == t)
        cm = confusion_matrix(y_true_bin, y_pred_bin, labels=[False, True])
        try:
            tpr[t] = cm[1,1] / (cm[1,1] + cm[1,0])
        except ZeroDivisionError:
            tpr[t] = np.nan
    return tpr
```

To evaluate fairness across laundering typologies, this block defines two key metrics:

---

#### 1. **Demographic Parity (DP)**

\[
DP_t = P(\hat{Y} = t \mid typology = t)
\]

- Measures whether each typology has an equal chance of being predicted as its true class.
- A high DP value implies the model frequently predicts the correct typology.

---

#### 2. **Equal Opportunity (EO)**

\[
EO_t = \frac{TP_t}{TP_t + FN_t}
\]

- Represents the True Positive Rate (TPR) for each typology.
- Focuses on **sensitivity**: how often the model correctly identifies a typology when it should.

---

These metrics will be applied to both the real-only model and the CGAN-augmented model to assess how data augmentation impacts **fairness across classes**.

This lays the groundwork for per-typology fairness comparison in upcoming blocks.


### Block 12: Fairness on Real-Only Model


```python
# Predictions and fairness on real-only model
y_pred_real = xgb_real.predict(X_real)

dp_real = demographic_parity(y_real, y_pred_real, y_real)
eo_real = equal_opportunity(y_real, y_pred_real, y_real)

fair_real_df = pd.DataFrame({
    'Typology': list(dp_real.keys()),
    'DP_Real': list(dp_real.values()),
    'EO_Real': list(eo_real.values())
}).sort_values(by='Typology')
```

This block applies the fairness metrics defined earlier to the baseline model trained only on real transaction data (`xgb_real`).

#### Process:
- The model makes predictions on the real dataset (`X_real`).
- Demographic Parity (DP) and Equal Opportunity (EO) are computed per typology.
- Results are stored in a DataFrame (`fair_real_df`) for later comparison.

**Columns:**
- `Typology`: Money laundering behavior code (class label)
- `DP_Real`: Demographic Parity score for the real-only model
- `EO_Real`: Equal Opportunity (TPR) for the real-only model

This baseline fairness distribution will serve as the reference when evaluating the fairness impact of CGAN-generated data.


### Block 13: Fairness on Real + CGAN Model

```python
# Predictions and fairness on CGAN-augmented model
y_pred_aug = xgb_aug.predict(X_aug)

dp_aug = demographic_parity(y_aug, y_pred_aug, y_aug)
eo_aug = equal_opportunity(y_aug, y_pred_aug, y_aug)

fair_aug_df = pd.DataFrame({
    'Typology': list(dp_aug.keys()),
    'DP_Aug': list(dp_aug.values()),
    'EO_Aug': list(eo_aug.values())
}).sort_values(by='Typology')
```

This block evaluates fairness for the model trained on the **augmented dataset** (`xgb_aug`), which includes both real and CGAN-generated synthetic transactions.

#### Steps:
- Predictions are made on the full augmented dataset (`X_aug`).
- Demographic Parity (`DP_Aug`) and Equal Opportunity (`EO_Aug`) are computed for each typology using the same definitions as in Block 10.
- Results are stored in `fair_aug_df`.

**Columns:**
- `Typology`: Laundering behavior class code
- `DP_Aug`: Demographic Parity score with CGAN-augmented training
- `EO_Aug`: Equal Opportunity score (TPR) with augmentation

These values will later be compared to their real-only counterparts to measure fairness drift introduced by synthetic augmentation.

```python
# Merge both into one DataFrame
fair_df = pd.merge(fair_real_df, fair_aug_df, on='Typology')

# Demographic Parity Plot
plt.figure(figsize=(10, 5))
plt.plot(fair_df['Typology'], fair_df['DP_Real'], marker='o', label='Real Only')
plt.plot(fair_df['Typology'], fair_df['DP_Aug'], marker='s', label='Real + CGAN')
plt.title("Demographic Parity by Typology")
plt.xlabel("Typology")
plt.ylabel("P(ŷ = t | typology = t)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{results_dir}/fairness_demographic_parity.png")
plt.close()

# Equal Opportunity Plot
plt.figure(figsize=(10, 5))
plt.plot(fair_df['Typology'], fair_df['EO_Real'], marker='o', label='Real Only')
plt.plot(fair_df['Typology'], fair_df['EO_Aug'], marker='s', label='Real + CGAN')
plt.title("Equal Opportunity by Typology")
plt.xlabel("Typology")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{results_dir}/fairness_equal_opportunity.png")
plt.show()
plt.close()
```

This block merges the fairness scores from both the real-only and CGAN-augmented models and plots their values side-by-side for each typology.

#### 1. **Demographic Parity Plot**

#### 2. **Equal Opportunity Plot**
- Measures True Positive Rate (TPR) per typology — i.e., sensitivity.

### Block 16: Fairness Bias Heatmap

```python
# Compute fairness gaps
fair_df['DP_Gap'] = fair_df['DP_Aug'] - fair_df['DP_Real']
fair_df['EO_Gap'] = fair_df['EO_Aug'] - fair_df['EO_Real']

# Heatmap of fairness gaps
plt.figure(figsize=(10, 5))
sns.heatmap(fair_df.set_index('Typology')[['DP_Gap', 'EO_Gap']], annot=True, cmap="coolwarm", center=0, fmt=".3f")
plt.title(" Fairness Gap Heatmap (Augmented - Real)")
plt.tight_layout()
plt.savefig(f"{results_dir}/fairness_gap_heatmap.png")
plt.show()
plt.close()
```

This block computes the **difference in fairness metrics** (Demographic Parity and Equal Opportunity) between the CGAN-augmented model and the real-only model:

\[
\text{DP\_Gap} = \text{DP}_{aug} - \text{DP}_{real}
\]
\[
\text{EO\_Gap} = \text{EO}_{aug} - \text{EO}_{real}
\]

A heatmap is plotted to highlight which typologies **benefit** (positive gap) or **suffer** (negative gap) in fairness as a result of CGAN-generated training data.

**Interpretation:**
- Typology `14` shows the largest positive DP gap, indicating improved representation after augmentation.
- Some typologies (e.g., `10`) exhibit negative EO gaps, highlighting fairness regressions to be addressed.

*Output:* `fairness_gap_heatmap.png`

This visual is central to understanding the fairness trade-offs introduced by generative augmentation — especially when modeling regulatory or compliance-sensitive domains like Anti-Money Laundering (AML).

### Block 19: SHAP–Fairness Correlation Analysis

```python
# Merge per-typology SHAP matrix with fairness gaps
shap_fair_df = typology_shap_matrix.copy()
shap_fair_df['Typology'] = shap_fair_df.index

# Join with fairness gap metrics
merged_typo_df = pd.merge(shap_fair_df, fair_df[['Typology', 'DP_Gap', 'EO_Gap']], on='Typology')

# Correlation: SHAP columns vs. fairness gap columns
correlation_matrix = merged_typo_df.drop(columns=['Typology']).corr()

# Extract only correlations between SHAP and DP/EO gaps
shap_cols = typology_shap_matrix.columns
fairness_corr = correlation_matrix.loc[shap_cols, ['DP_Gap', 'EO_Gap']]

# Plot correlation heatmap
plt.figure(figsize=(8, len(shap_cols)*0.4))
sns.heatmap(fairness_corr, annot=True, cmap="coolwarm", center=0, fmt=".2f")
plt.title("SHAP vs. Fairness Gap Correlation (per Feature)")
plt.tight_layout()
plt.savefig(f"{results_dir}/shap_fairness_correlation.png")
plt.show()
plt.close()
```

To assess whether changes in explainability contribute to fairness drift, this block computes the **Pearson correlation** between:

- Each feature’s average SHAP value per typology  
- Corresponding fairness gaps (`DP_Gap`, `EO_Gap`) per typology

This helps identify which features most strongly influence fairness shifts after CGAN-based augmentation.

#### Key Observations:
- `Amount` is **negatively correlated** with both DP and EO gaps, suggesting fairness may degrade when this feature dominates.
- `Sender_bank_location` and `Receiver_bank_location` show mild positive correlation with fairness — possibly stabilizing influence.

*Output:* `shap_fairness_correlation.png`

This block strengthens model auditability by linking interpretability signals with fairness outcomes — valuable for responsible AI deployment.
